````python
import re
import json
````

# Calculator Tool

The calculator tool extracts a mathematical expression from the user's query and safely evaluates it.

Example:

Input:

calculate 25 + 15

Output:

{
    "type":"calculation",
    "result":40
}

````python
def calculator(expression):
    """
    Evaluates a mathematical expression safely.
    """

    try:
        allowed = "0123456789+-*/(). "

        for ch in expression:
            if ch not in allowed:
                raise ValueError("Invalid character found.")

        result = eval(expression)

        return {
            "type": "calculation",
            "result": result
        }

    except Exception as e:

        return {
            "type": "error",
            "result": str(e)
        }
````

# Keyword Extraction Tool

This tool extracts important words from the user's text after removing common stopwords.

Example:

Input

keywords Artificial Intelligence is transforming education rapidly.

Output

{
    "type":"keywords",
    "result":[
        "artificial",
        "intelligence",
        "transforming",
        "education",
        "rapidly"
    ]
}

````python
def extract_keywords(text):

    stop_words = {
        "the","is","a","an","to","of","for",
        "and","or","in","on","with","this",
        "that","it","are","was","were","be",
        "as","at","by","from","about","into"
    }

    words = re.findall(r"[A-Za-z]+", text.lower())

    keywords = []

    for word in words:
        if word not in stop_words:
            keywords.append(word)

    return {
        "type": "keywords",
        "result": keywords
    }
````

# Agent Pipeline

The agent acts as the central controller of the workflow.

Routing Logic:

- If the query contains **calculate**, route to Calculator Tool.
- If the query contains **keywords**, route to Keyword Extraction Tool.
- Otherwise, return a general response.
- Invalid inputs are handled using structured error messages.

Every output follows the JSON format:

{
    "type":"...",
    "result":...
}

# Agent Function

The `agent(query)` function is the central controller of the pipeline.

It performs the following tasks:

1. Validates user input.
2. Converts the query to lowercase for consistent matching.
3. Routes the query based on keywords:
   - **calculate** → Calculator Tool
   - **keywords** → Keyword Extraction Tool
   - otherwise → General Response
4. Handles invalid inputs safely.
5. Returns structured JSON output.

This demonstrates a simple rule-based single-agent pipeline.

````python
def agent(query):
    """
    Main Agent Function
    Routes the query to the correct tool.
    """

    print("\n==============================")
    print("Agent Received Query:")
    print(query)
    print("==============================")

    try:

        # Input Validation
        if not isinstance(query, str):
            raise TypeError("Input must be a string.")

        query = query.strip()

        if query == "":
            raise ValueError("Query cannot be empty.")

        lower_query = query.lower()

        print("Validation Passed ✓")
        print("Routing Decision...")

        # ---------------------------------
        # Calculator Routing
        # ---------------------------------

        if "calculate" in lower_query:

            expression = lower_query.replace("calculate", "").strip()

            print("→ Routed to Calculator Tool")

            return calculator(expression)

        # ---------------------------------
        # Keyword Extraction Routing
        # ---------------------------------

        elif "keywords" in lower_query:

            text = lower_query.replace("keywords", "").strip()

            print("→ Routed to Keyword Extraction Tool")

            return extract_keywords(text)

        # ---------------------------------
        # General Response
        # ---------------------------------

        else:

            print("→ Routed to General Response")

            return {
                "type": "general",
                "result": f"You said: {query}"
            }

    except Exception as e:

        print("Routing Failed ✗")

        return {
            "type": "error",
            "result": str(e)
        }
````

# Interactive Agent Pipeline
